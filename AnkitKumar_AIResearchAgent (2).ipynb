```python
from openai import OpenAI

client = OpenAI(api_key="API KEY")

# Step 1: User input
topic = input("Enter your research topic: ")

# Step 2: Generate short sub-questions
sub_qs_prompt = f"Generate exactly 3 simple and short research sub-questions (1 line each) based on the topic: {topic}. Do NOT add explanation."
sub_questions = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": sub_qs_prompt}]
)

sub_qs_text = sub_questions.choices[0].message.content
print("\nGenerated Sub-questions:\n", sub_qs_text)

# Step 3: Short summary
summary_prompt = f"Write a short 2-3 line summary of these sub-questions:\n{sub_qs_text}"
summary = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": summary_prompt}]
)

print("\nSummary:\n", summary.choices[0].message.content)
```
